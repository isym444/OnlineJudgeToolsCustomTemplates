Efficient 2 pointer approach to solve number of distinct numbers in a range

O(N x sqrt(N));

https://atcoder.jp/contests/awc0015/tasks/awc0015_e


```cpp
int main()
{
  nyan;
  int n, q;
  cin >> n >> q;
  vector<int> p(n);
  for (int i = 0; i < n; i++)
  {
    cin >> p[i];
    p[i]--;
  }
  struct Query
  {
    int l, r, idx;
  };
  vector<Query> queries(q);
  for (int i = 0; i < q; i++)
  {
    cin >> queries[i].l >> queries[i].r;
    queries[i].l--;
    queries[i].idx = i;
  }
  int B = sqrt(n);
  sort(queries.begin(), queries.end(), [&](const Query &a, const Query &b)
       {
        int ba = a.l / B, bb = b.l / B; // into bucket based on position of l
        if (ba != bb) return ba < bb; // sort by those in more left buckets
        return (ba & 1) ? a.r > b.r : a.r < b.r; }); // then by zigzag asc/desc r -> not necessary but optimization
  vector<int> cnt(n, 0);
  int distinct = 0;

  auto add = [&](int i)
  {
    cnt[p[i]] = cnt[p[i]] + 1;
    if (cnt[p[i]] == 1) // was 0 before, now 1 → new species
      distinct++;
  };

  auto rem = [&](int i)
  {
    cnt[p[i]] = cnt[p[i]] - 1;
    if (cnt[p[i]] == 0) // just hit 0 → species gone
      distinct--;
  };

  vector<int> ans(q);
  int l = 0, r = 0;
  for (auto &[ql, qr, idx] : queries)
  {
    while (l > ql)
    {
      l = l - 1;
      add(l);
    }
    while (l < ql)
    {
      rem(l);
      l = l + 1;
    }
    while (r < qr)
    {
      add(r);
      r = r + 1;
    }
    while (r > qr)
    {
      r = r - 1;
      rem(r);
    }
    ans[idx] = distinct;
  }
  for (int i = 0; i < q; i++)
    cout << ans[i] << endl;
}
```
