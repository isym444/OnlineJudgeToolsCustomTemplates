x is the initial node node
dist[targ] holds the shortest path from x to any targ node



```cpp
vector<vector<pair<ll, ll>>> G(N); // target node, edge-weight
priority_queue<pair<ll, ll>> qq; // -ve distance to node, node
vll distance(N + 5, INF);
vll processed(N + 5);
distance[x] = 0;
q.push({0, x});
while (!q.empty())
{
  int a = q.top().second;
  q.pop();
  if (processed[a])
    continue;
  processed[a] = true;
  for (auto u : G[a])
  {
    int b = u.first, w = u.second;
    if (distance[a] + w < distance[b])
    {
      distance[b] = distance[a] + w;
      q.push({-distance[b], b});
    }
  }
}
```
